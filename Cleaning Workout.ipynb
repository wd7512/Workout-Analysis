```python
f = open('Legs.txt')
raw = f.readlines()
f.close()

while '\n' in raw:
    raw.remove('\n')
    
raw = [line.strip() for line in raw]
        
raw
```

```python

days = []
new_day = False
date = None
workout = []
for line in raw:
    if 'â€”' in line:
        new_day = True
        for w in workout:
            
            days.append([date,w])
    
    if new_day == True:
        
        workout = []
        slash = line.index('/')
        date = line[:slash+3]
        new_day = False
    else:
        workout.append(line)
        
days = days[:]

for row in days:
    print(row)
```

```python
def homo(strng):
    if ('iso' in strng) and ('leg' in strng):
        print('ok')
        return 'ISO LEG PRESS'
    else:
        return strng
    
for val in days:
    val[1] = homo(val[1])
    

days
```
